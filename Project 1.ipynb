# REASONS WHY I USED THE MEAN METHOD

#### The mean method represents the average of a set of data, meaning it is neither the smallest nor the largest nor the middle value.

#### In the case of red wine ,the mean method is most appropriate because most of the chemical compostion of the red wine for example the Sulphur dioxide .If a winemaker uses too much SO2, it can kill the “good” yeast, haulting fermentation before the desired end point. It can also stop malolactic fermentation from completing, yield a red wine which has an unfinished taste .If the wine maker uses very small amount of SO2 ,the benefit, regulating the growth of harmful yeast and bacterial growth in the wine to prevent fermentation, that is obtained by using the standard amount of SO2 may not be maximized.

#### This also applies to other features such as the fixed acidity,pH ,chlorides amongst other, which must have an average value to replace null values in order to maintain their usefulness within the red wine.

```python
Qm=red_wine_data.quality.mean()
pHm=red_wine_data.pH.mean()
Dm=red_wine_data.density.mean()
FreeSO2m=red_wine_data.free_sulfur_dioxide.mean()
TotalSO2m=red_wine_data.total_sulfur_dioxide.mean()
Cm=red_wine_data.chlorides.mean()
CAm=red_wine_data.citric_acid.mean()
FAm=red_wine_data.fixed_acidity.mean()
```

```python
red_wine_data['quality']=red_wine_data['quality'].fillna(Qm)
red_wine_data['pH']=red_wine_data['pH'].fillna(pHm)
red_wine_data['density']=red_wine_data['density'].fillna(Dm)
red_wine_data['free_sulfur_dioxide']=red_wine_data['free_sulfur_dioxide'].fillna(FreeSO2m)
red_wine_data['total_sulfur_dioxide']=red_wine_data['total_sulfur_dioxide'].fillna(TotalSO2m)
red_wine_data['chlorides']=red_wine_data['chlorides'].fillna(Cm)
red_wine_data['citric_acid']=red_wine_data['citric_acid'].fillna(CAm)
red_wine_data['fixed_acidity']=red_wine_data['fixed_acidity'].fillna(FAm)
```

## Residual Sugar
---
### One element impacting how wine tastes and feels in the mouth is residual sugar (RS.) Residual sugar represents the amount of sugar remaining after alcoholic fermentation, when yeasts have converted grape sugars into alcohol. The higher the level of residual sugar ,more than 4g/L, the sweeter the wine and the fuller or more viscous it can taste in the mouth.Owing to this, wines with higher levels of residual sugar sell more on the market.On the contrary, a residual sugar level of more than 2g/L may be unsuitable for diabetic persons.
